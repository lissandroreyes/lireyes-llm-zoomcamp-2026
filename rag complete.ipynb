```python
from dotenv import load_dotenv
load_dotenv()

from ingest import load_faq_data, build_index
from rag_helper import RAGBase
from google import genai  # Importamos la librería nativa de Google

# 1. Cargamos datos y construimos el índice de búsqueda
documents = load_faq_data()
index = build_index(documents)

# 2. Inicializamos el cliente oficial de Gemini
gemini_client = genai.Client()

# 3. Instanciamos la clase pasando nuestro cliente de Google
assistant = RAGBase(
    index=index,
    llm_client=gemini_client,
)

# 4. Probamos el flujo RAG modularizado
answer = assistant.rag("I just discovered the course. Can I join now?")
print(answer)
```

```python
# 1. Definimos las nuevas instrucciones personalizadas (más estrictas)
custom_instructions = """
You're a course teaching assistant.
Answer the QUESTION based on the CONTEXT from the FAQ database.
Use only the facts from the CONTEXT when answering the QUESTION.
""".strip()

# 2. Creamos un nuevo asistente pasando tu cliente de Gemini y las nuevas reglas
assistant_estricto = RAGBase(
    index=index,
    llm_client=gemini_client,       # Tu cliente de Google
    instructions=custom_instructions, # Sobrescribe las instrucciones por defecto
)

# 3. Hacemos la prueba
print(assistant_estricto.rag("How do I get a certificate?"))
```
