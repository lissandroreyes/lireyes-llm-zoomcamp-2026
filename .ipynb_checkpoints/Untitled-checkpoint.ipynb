```python
#Esta sección del curso te muestra de forma visual por qué la función rag que escribimos dentro de nuestra clase RAGBase
#es una estructura estática o lineal (un pipeline fijo).El código se ejecuta estrictamente de arriba hacia abajo 
#sin importar lo que pase en el camino

def rag(self, query):
    search_results = self.search(query)             # 1. Busca pase lo que pase
    prompt = self.build_prompt(query, search_results) # 2. Construye el prompt (aunque esté vacío)
    answer = self.llm(prompt)                        # 3. Llama a Gemini con lo que tenga
    return answer
```

```python
#Asking without tools
#First, let's see what the LLM does without any tools. We ask it a course-specific question and look at the answer.

# 1. Definimos la pregunta específica del curso
prompt_usuario = "I just discovered the course. Can I join it?"

# 2. Enviamos la petición directamente a Gemini sin pasarle contexto ni herramientas
response = gemini_client.models.generate_content(
    #model='gemini-2.5-flash',
    model='gemini-3.5-flash', 
    contents=prompt_usuario,
)

# 3. Obtenemos el texto de la respuesta (equivalente a response.output_text)
print("Respuesta de Gemini sin herramientas:")
print(response.text)

#The model answers from its general knowledge, something like "it depends on the course" or "check the course website". 
#It doesn't know about our FAQ, so the answer is vague and not helpful. 
#This is exactly why we need RAG, and why we want to hand the model a tool.

```

```python
#Defining the tool
#First we define a top-level search function that queries the index directly. 
#The model will reference it by this name. We keep the Python function and the tool name aligned so the dispatch is easier later.

# Definición de la función de búsqueda local
def search(query):
    boost_dict = {"question": 3.0, "section": 0.5}
    filter_dict = {"course": "llm-zoomcamp"}

    return index.search(
        query,
        num_results=5,
        boost_dict=boost_dict,
        filter_dict=filter_dict
    )
```

```python
#we're just making an HTTP call, so we describe the tool in JSON rather than in Python. The same schema would work from TypeScript or Java.

# Definición del esquema JSON adaptado para Gemini
search_tool = {
    "function_declarations": [
        {
            "name": "search",
            "description": "Search the FAQ database for entries matching the given query.",
            "parameters": {
                "type": "OBJECT",
                "properties": {
                    "query": {
                        "type": "STRING",
                        "description": "Search query text to look up in the course FAQ."
                    }
                },
                "required": ["query"]
            }
        }
    ]
}


#The description is the most important field, because the model reads it to decide when to call the function. 
#parameters is a JSON schema for the arguments, and we mark query as required so the model always fills it in.
```

```python
# Historial de conversación inicial
messages = [
    {
        "role": "user", 
        "parts": [{"text": "How do I run Olama locally?"}]
    }
]
```

```python
# 2. Enviamos la pregunta incluyendo la herramienta en la configuración
#Sending the question with the tool
# Primera llamada a Gemini
response = gemini_client.models.generate_content(
    model='gemini-3.5-flash',
    contents=messages,
    config={
        'tools': [search_tool]
    }
)

# Guardamos la llamada a la función para usarla en el siguiente bloque
# Equivalente a response.output del curso
print("Llamada detectada:")
print(response.function_calls)
```

```python
# Muestra las llamadas a funciones generadas por el modelo, equivalente a response.output
print(response.function_calls)
```

```python
#Executing the function and sending the result back
#3. Ejecutar la función e integrarla al historialA diferencia de OpenAI que requiere librerías como json.loads, 
#Gemini nos da los argumentos ya procesados en un diccionario nativo de Python dentro de response.function_calls[0].args.

# 1. Extraemos la llamada y los argumentos directamente
call = response.function_calls[0]
args = call.args  # Gemini ya te da un diccionario, no necesitas json.loads

# 2. Ejecutamos nuestra función Python 'search' usando desempaquetado de diccionario
results = search(**args)

# 3. Convertimos el resultado a JSON string tal como pide el curso
import json
result_json = json.dumps(results, indent=2)

# 4. Actualizamos el historial de mensajes al estilo Gemini:
# Primero, pasamos la respuesta completa que nos dio el modelo (para que vea su propia llamada)
messages.append(response.candidates[0].content)

# Segundo, añadimos la respuesta de la función simulando la estructura que espera la API
messages.append({
    "role": "user",
    "parts": [{
        "function_response": {
            "name": call.name,
            "response": {"result": result_json}
        }
    }]
})
```

```python
#Asking the model again
# Segunda llamada con todo el historial de la conversación
#Sending the question with the tool
second_response = gemini_client.models.generate_content(
    model='gemini-2.5-flash',
    contents=messages,
    config={
        'tools': [search_tool]
    }
)

# Imprimimos la respuesta final enriquecida con RAG Agéntico
# Equivalente a response.output_text del curso
print("Respuesta final del Agente:")
print(second_response.text)
```

```python
#Token usage and cost
# Extraemos los tokens del segundo objeto de respuesta de Google
usage = second_response.usage_metadata
input_tokens = usage.prompt_token_count
output_tokens = usage.candidates_token_count

def calculate_gemini_price(input_tokens, output_tokens):
    # Precios de referencia por millón de tokens para la familia Flash de Gemini
    INPUT_PRICE_PER_MILLION = 0.30
    OUTPUT_PRICE_PER_MILLION = 1.20

    input_cost = (input_tokens / 1_000_000) * INPUT_PRICE_PER_MILLION
    output_cost = (output_tokens / 1_000_000) * OUTPUT_PRICE_PER_MILLION
    total_cost = input_cost + output_cost

    return {
        "input_cost": input_cost,
        "output_cost": output_cost,
        "total_cost": total_cost,
    }

result = calculate_gemini_price(input_tokens, output_tokens)
print(f"Tokens de Entrada: {input_tokens} | Tokens de Salida: {output_tokens}")
print("Costo de la segunda llamada: $", round(result["total_cost"], 8))
```

```python
#A function-call helper
import json

def make_call(call):
    # Gemini ya provee los argumentos en un diccionario nativo (call.args)
    args = call.args

    if call.name == "search":
        result = search(**args)

    result_json = json.dumps(result, indent=2)

    # Devolvemos la estructura exacta esperada por la API de Gemini
    return {
        "role": "user",
        "parts": [{
            "function_response": {
                "name": call.name,
                "response": {"result": result_json}
            }
        }]
    }
```

```python
question = "I just discovered the course. Can I join it?"

# El equivalente al diseño de mensajes del curso
messages = [
    {
        "role": "user", 
        "parts": [{"text": question}]
    }
]

# El equivalente a la llamada responses.create con herramientas
response = gemini_client.models.generate_content(
    model='gemini-3.5-flash',
    contents=messages,
    config={
        'tools': [search_tool],
        'system_instruction': instructions # Aquí se pasa la instrucción del desarrollador
    }
)

# El equivalente a messages.extend(response.output)
if response.candidates and response.candidates[0].content:
    messages.append(response.candidates[0].content)

has_function_calls = False

# El equivalente al bucle 'for item in response.output' del curso
if response.function_calls:
    for call in response.function_calls:
        print("function_call:", call.name, call.args)
        call_output = make_call(call)
        messages.append(call_output)
        has_function_calls = True

# El equivalente al 'elif item.type == "message"' del curso
# Agregamos la verificación para evitar el aviso (Warning) de Google
elif response.text:
    print("ASSISTANT:")
    print(response.text)
```

```python
it = 1

while True:
    print(f"iteration #{it}...")
    has_function_calls = False

    # Llamada a la API nativa de Gemini en cada vuelta del bucle
    response = gemini_client.models.generate_content(
        model='gemini-3.5-flash',
        contents=messages,
        config={
            'tools': [search_tool],
            'system_instruction': instructions
        }
    )

    # Equivalente a messages.extend(response.output)
    if response.candidates and response.candidates[0].content:
        messages.append(response.candidates[0].content)

    # Equivalente a evaluar las funciones dentro del bucle for del curso
    if response.function_calls:
        for call in response.function_calls:
            print("function_call:", call.name, call.args)
            call_output = make_call(call)
            messages.append(call_output)
            has_function_calls = True

    # Equivalente al 'elif item.type == "message"' del curso
    elif response.text:
        print("ASSISTANT:")
        print(response.text)

    it = it + 1
    
    # Condición de salida idéntica: si la IA no pide herramientas, rompemos el bucle
    if has_function_calls == False:
        break
```

```python
instructions = """
You're a course teaching assistant.
You're given a question from a course student and your task is to answer it.

If you want to look up information, use the search function. 
Use as many keywords from the user question as possible when making first requests.

CRITICAL: If a search query yields poor or zero results, analyze if a technical term or tool name might have a typo or spelling mistake (for example, "Olama" instead of "Ollama"). Correct the spelling and try a new search immediately with the corrected term.

Make multiple searches.

Try to expand your search by using new keywords
based on the results you get from the search.

At the end, ask if there are other areas that the user wants to explore.
""".strip()
```

```python
#Wrapping it in a function
def agent_loop(instructions, question, model="gemini-2.5-flash") -> str:
    # 1. El equivalente al diseño de mensajes del curso
    messages = [
        {
            "role": "user", 
            "parts": [{"text": question}]
        }
    ]

    it = 1
    last_answer = ""

    while True:
        print(f"iteration #{it}...")
        has_function_calls = False

        # 2. El equivalente a la llamada responses.create con herramientas
        response = gemini_client.models.generate_content(
            model=model,
            contents=messages,
            config={
                'tools': [search_tool],
                'system_instruction': instructions # Rol developer/system mapeado aquí
            }
        )

        # 3. El equivalente a messages.extend(response.output)
        if response.candidates and response.candidates[0].content:
            messages.append(response.candidates[0].content)

        # 4. El equivalente al 'if item.type == "function_call"' del curso
        if response.function_calls:
            for call in response.function_calls:
                print("function_call:", call.name, call.args)
                call_output = make_call(call)
                messages.append(call_output)
                has_function_calls = True

        # 5. El equivalente al 'elif item.type == "message"' del curso
        elif response.text:
            print("ASSISTANT:")
            last_answer = response.text
            print(last_answer)

        it = it + 1
        
        # Condición de salida idéntica
        if has_function_calls == False:
            break

    return last_answer
```

```python
import os
import json
from dotenv import load_dotenv
from google import genai  
from toyaikit.tools import Tools
from toyaikit.chat import IPythonChatInterface
from toyaikit.chat.runners import OpenAIResponsesRunner, DisplayingRunnerCallback

# 1. Cargamos las variables de tu archivo .env
load_dotenv()

# 2. Inicializamos tu cliente oficial de Google Gemini
gemini_native_client = genai.Client()

# 3. Conector personalizado independiente de OpenAI
class CustomGeminiToyAIClient:
    def __init__(self, model="gemini-2.5-flash-lite"):
        self.model = model

    def send_request(self, chat_messages, tools, output_format=None):
        # Extraemos el último prompt enviado por el framework
        user_prompt = "How do I run Ollama locally?"
        if chat_messages:
            if 'prompt' in chat_messages[-1]:
                user_prompt = chat_messages[-1]['prompt']
            elif 'input' in chat_messages[-1]:
                for item in chat_messages[-1]['input']:
                    if item.get('type') == 'message':
                        user_prompt = item['content']['text']

        # Llamada 100% nativa con la librería de Google
        response = gemini_native_client.models.generate_content(
            model=self.model,
            contents=user_prompt,
            config={
                'tools': [search_tool],  
                'system_instruction': instructions
            }
        )

        # Clases internas simuladas para emular el formato de OpenAI
        class MockOutput:
            def __init__(self, res):
                self.type = "function_call" if res.function_calls else "message"
                if self.type == "function_call":
                    # Si viene como lista, tomamos el primer elemento
                    call = res.function_calls[0] if isinstance(res.function_calls, list) else res.function_calls
                    self.name = call.name
                    self.arguments = json.dumps(call.args) if hasattr(call, 'args') else "{}"
                    self.call_id = "call_123"
                else:
                    class ContentItem:
                        text = res.text or ""
                    self.content = [ContentItem()]

        # CORRECCIÓN: Estructura de tokens mapeada al formato que toyaikit espera
        class MockUsage:
            def __init__(self, meta):
                # Traducimos los nombres de variables de Google a los de OpenAI
                self.input_tokens = meta.prompt_token_count if meta else 0
                self.output_tokens = meta.candidates_token_count if meta else 0

        class MockResponse:
            output = [MockOutput(response)]
            output_text = response.text or ""
            usage = MockUsage(response.usage_metadata if hasattr(response, 'usage_metadata') else None)

        return MockResponse()

# 4. Instanciamos el cliente usando el conector corregido
llm_client_gemini = CustomGeminiToyAIClient(model="gemini-3.5-flash")

# 5. Registramos las herramientas
agent_tools = Tools()
agent_tools.add_tool(search, search_tool)

# 6. Inicializamos la interfaz de chat interactiva de Jupyter
chat_interface = IPythonChatInterface()
callback = DisplayingRunnerCallback(chat_interface)

# 7. Creamos tu Runner oficial del curso
runner = OpenAIResponsesRunner(
    tools=agent_tools,
    developer_prompt=instructions,
    chat_interface=chat_interface,
    llm_client=llm_client_gemini  
)
```

```python
# Hacemos la primera pregunta para inicializar el objeto 'result'
result = runner.loop(
    prompt="How do I run Ollama locally?",
    callback=callback,
)
```
